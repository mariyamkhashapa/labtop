```python
#preprocessing 
df.drop(columns=["Unnamed: 0"],inplace=True)
print(df.head())
```

```python
#delete GB from ram and kg from weight
df['Ram'] = df['Ram'].str.replace('GB','')
df['Weight'] = df['Weight'].str.replace('kg','')
df['Ram'] = df['Ram'].astype('int32')
df['Weight'] = df['Weight'].astype('float32')
print(df.head())
df.info()
```

```python
new = df['ScreenResolution'].str.split('x',n=1,expand=True)
df['X_res'] = new[0]
df['Y_res'] = new[1]
print(df.sample(5))
```

```python
df['X_res'] = df['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0]) 
print(df.head())
```

```python
df['X_res'] = df['X_res'].astype('int')
df['Y_res'] = df['Y_res'].astype('int')
print(df.info())
```

```python

df['ppi'] = (((df['X_res']**2) + (df['Y_res']**2))**0.5/df['Inches']).astype('float') #function to represent pixel per inch new feature
```

```python
df.corr()['Price']
```

```python
df.drop(columns=['ScreenResolution'],inplace=True)
print(df.head())
```

```python
df.drop(columns=['Inches','X_res','Y_res'],inplace=True)
print(df.head())
```

```python
df['Cpu'].value_counts()
df['Cpu Name'] = df['Cpu'].apply(lambda x:" ".join(x.split()[0:3])) #extract cpu name as new feature
print(df.head())
```

```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['Cpu brand'] = df['Cpu Name'].apply(fetch_processor)            
print(df.head())
```

```python
df['Cpu brand'].value_counts().plot(kind='bar')
```

```python
sns.barplot(x=df['Cpu brand'],y=df['Price'])
plt.xticks(rotation='vertical')
plt.show()
```
